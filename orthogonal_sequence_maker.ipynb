```python
def cc_round(main_scaffold, artificial_scaffold, padding=0, strides=1):
    corr = (len(main_scaffold) - len(artificial_scaffold) + 1) * [0]
    for l in range(len(corr)):
        counter = 0
        for i in range(len(artificial_scaffold)):
            if main_scaffold[l+i] == artificial_scaffold[i]:
                counter += 1
        corr[l] = [counter, main_scaffold[l:(l+len(artificial_scaffold))%len(main_scaffold)]]
    return corr
```

```python
def substitution_score(corr_mat, main_scaffold, artificial_scaffold, choice, i):
    artificial_scaffold = artificial_scaffold[:i] + choice  + artificial_scaffold[i+1:]
    new_corr_mat = cc_round(main_scaffold, artificial_scaffold)
    loss = 0
    for i in range(len(new_corr_mat)):
        row_loss = np.exp(new_corr_mat[i][0]) - np.exp(corr_mat[i][0])
        loss += row_loss
    return loss
```

```python
def generate_random_sequence(size):
    alphabet = ["A", "T", "C", "G"]
    sequence = "".join(list(np.random.choice(alphabet, size=size)))
    return sequence
```

```python
def generate_artificial_scaffold(size=60, rounds=1):

    seed_scaffold = generate_random_sequence(size)
    convolved_scaffold = seed_scaffold
    prev_stack_top_idx = -1
    curr_stack_top_idx = 0
    prev_sub_group = [None, None]
    curr_sub_group = [None, None]
    for j in range(rounds):
        corr_mat = cc_round(main_scaffold, convolved_scaffold)
        corr_arr = list(map(lambda x: x[0], corr_mat))
        corr_mat_sorted = sorted(corr_mat, key=lambda x: x[0], reverse=True)
        stack_change = (curr_stack_top_idx-prev_stack_top_idx != 0)
        if j == 0 or stack_change:
            prev_stack_top_idx += 1

        most_similar_partition_val = corr_mat_sorted[curr_stack_top_idx][0]

        most_similar_partition = corr_mat_sorted[curr_stack_top_idx][1]
        sub_score_matrix = []
        for i in range(len(convolved_scaffold)):
            if convolved_scaffold[i] != most_similar_partition[i]:
                continue
            choices, choice_scores = [], {}
            if convolved_scaffold[i] == "A":
                choices = ["T", "C", "G"]
            elif convolved_scaffold[i] == "T":
                choices = ["A", "C", "G"]
            elif convolved_scaffold[i] == "G":
                choices = ["A", "T", "C"]
            elif convolved_scaffold[i] == "C":
                choices = ["A", "T", "G"]
            for choice in choices:
                choice_score = substitution_score(corr_mat, main_scaffold, convolved_scaffold, choice, i)
                choice_scores[choice] = choice_score


            best_sub_choice = min(choice_scores, key=choice_scores.get)
            sub_score_matrix.append([i, best_sub_choice, choice_scores[best_sub_choice]])

        min_sub_value = float('inf')
        min_sub_choice = [None, None]
        for idx, sub_choice, sub_value in sub_score_matrix:
            if sub_value < min_sub_value:
                min_sub_value = sub_value
                min_sub_choice = [idx, sub_choice]
        if prev_sub_group[0] == min_sub_choice[0] and prev_sub_group[1] == min_sub_choice[1]:
            curr_stack_top_idx += 1
        else:
            prev_sub_group = curr_sub_group
            curr_sub_group = min_sub_choice
            convolved_scaffold = convolved_scaffold[:min_sub_choice[0]] \
            + min_sub_choice[1] + convolved_scaffold[min_sub_choice[0]+1:]

        if j == rounds-1:
            corr_mat = cc_round(main_scaffold, convolved_scaffold)
            corr_mat_sorted = sorted(corr_mat, key=lambda x: x[0], reverse=True)
            corr_arr = list(map(lambda x: x[0], corr_mat))
#             print("Mean: ", np.mean(corr_arr), \
#               "Median: ", np.median(corr_arr), \
#               "Std: ", np.std(corr_arr), \
#               "Var: ", np.var(corr_arr), \
#               "Min: ", np.min(corr_arr), \
#               "Max: ", np.max(corr_arr))
            print("Generated sequnce: ", convolved_scaffold) #, "\n", \
#                   "1st most similar: ", corr_mat_sorted[0], "\n", \
#                   "2nd most similar: ", corr_mat_sorted[1], "\n", \
#                   "3rd most similar: ", corr_mat_sorted[2], "\n")

    return convolved_scaffold
```

```python
def generate_artificial_scaffold_library(size, count=1, rounds=1):
    scaffold_lib = []
    for item in range(count):
        new_scaffold = generate_artificial_scaffold(size, rounds)
        scaffold_lib.append(new_scaffold)
    return scaffold_lib
```

```python
def levenshtein_distance(s, t):
    m, n = len(s), len(t)
    dp = [[0] * (n+1) for _ in range(m+1)]

    for i in range(m+1):
        dp[i][0] = i
    for j in range(n+1):
        dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            if s[i-1] == t[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) + 1

    return dp[m][n]
```

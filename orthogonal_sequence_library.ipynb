```python
def cc_round(main_scaffold, artificial_scaffold, padding=0, strides=1):
    corr = (len(main_scaffold) - len(artificial_scaffold) + 1) * [0]
    for l in range(len(corr)):
        counter = 0
        for i in range(len(artificial_scaffold)):
            if main_scaffold[l+i] == artificial_scaffold[i]:
                counter += 1
        corr[l] = [counter, main_scaffold[l:(l+len(artificial_scaffold))%len(main_scaffold)]]
#     max_val = max(corr)
#     max_index = corr.index(max_val)
    return corr #max_val, substitute_sequence(main_scaffold[max_index:max_index+len(artificial_scaffold)], artificial_scaffold, 5)
```

```python
def generate_artificial_scaffold_library(size=60, count=1, rounds=1):
    scaffold_lib = []
    seed_lib = []
    for i in range(count):
        seed_scaffold = generate_random_sequence(size)
        convolved_scaffold = seed_scaffold
        for j in range(rounds):
            corr_mat = cc_round(main_scaffold, convolved_scaffold)
            if i == 0:
                plt.figure()
                plt.hist(list(map(lambda x: x[0], corr_mat)), density=True)
                plt.show()
            corr_mat = sorted(corr_mat, key=lambda x: x[0], reverse=True)
            most_similar_partition = corr_mat[0][1]
            for i in range(len(convolved_scaffold)):
                if convolved_scaffold[i] != most_similar_partition[i]:
                    continue
                choices, choice_scores = [], {}
                if convolved_scaffold[i] == "A":
                    choices = ["T", "C", "G"]
                elif convolved_scaffold[i] == "T":
                    choices = ["A", "C", "G"]
                elif convolved_scaffold[i] == "G":
                    choices = ["A", "T", "C"]
                elif convolved_scaffold[i] == "C":
                    choices = ["A", "T", "G"]
                for choice in choices:
                    choice_score = substitution_score(corr_mat, main_scaffold, convolved_scaffold, choice, i)
                    choice_scores[choice] = choice_score
                
                best_sub_choice = min(choice_scores, key=choice_scores.get)
#                 print(best_sub_choice)
                convolved_scaffold = convolved_scaffold[:i] + best_sub_choice + convolved_scaffold[i+1:]
                    
        corr_mat = cc_round(main_scaffold, convolved_scaffold)
#         if i == rounds-1:
        plt.figure()
        plt.hist(list(map(lambda x: x[0], corr_mat)), density=True)
        plt.show()

        scaffold_lib.append(convolved_scaffold)
        seed_lib.append(seed_scaffold)
    return seed_lib, scaffold_lib

def substitution_score(corr_mat, main_scaffold, artificial_scaffold, choice, i):
    artificial_scaffold = artificial_scaffold[:i] + choice  + artificial_scaffold[i+1:]
    new_corr_mat = cc_round(main_scaffold, artificial_scaffold)
    loss = 0
    for row in new_corr_mat:
        loss += -np.log(row[0])
#     print(loss)
    return loss
```

```python
def generate_random_sequence(size):
    alphabet = ["A", "T", "C", "G"]
    sequence = "".join(list(np.random.choice(alphabet, size=size)))
    return sequence
```

```python
def substitute_sequence(primary, secondary, threshold):
    replacement_scheme = {"A": ["G", "C"], "T": ["G", "C"], "G": ["A", "T"], "C": ["A", "T"]}
    matching_seq_list = []
    replaced = 0
    for i in range(len(primary)):
        if primary[i] == secondary[i]:
            matching_seq_list.append(i)
    np.random.shuffle(matching_seq_list)
    
    for index in matching_seq_list:
        if replaced > threshold:
            break
        secondary = secondary[:index] + np.random.choice(replacement_scheme[secondary[index]])[0] + \
        secondary[index+1:]
        replaced += 1
    return secondary
```
